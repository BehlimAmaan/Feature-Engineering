```python
import pandas as pd
import numpy as np 
import sklearn
```

```python
n_sample= 10000
class_0_ratio = 0.9

class_0 = int(n_sample * 0.9)
class_1= int(n_sample - class_0)
```

```python
class_0, class_1
```

```python
Class_Majority = pd.DataFrame({
    'Feature_1': np.random.normal(loc=0, scale=1, size=class_0),
    'Feature_2': np.random.normal(loc=0, scale=1,size =class_0),
    'Target': [0] * class_0
})

Class_Minority = pd.DataFrame({
    'Feature_1': np.random.normal(loc=2, scale=1, size=class_1),
    'Feature_2': np.random.normal(loc=2, scale=1,size =class_1),
    'Target': [1] * class_1
})

```

```python
df = pd.concat([Class_Majority, Class_Minority]).reset_index(drop=True)
```

```python
df
```

```python
df['Target'].value_counts()
```

### UpSampling 

```python
df_minority = df[df['Target']==1]
df_majority = df[df['Target']==0]
```

```python
df_minority
```

```python
from sklearn.utils import resample
df_minority_updated=resample(df_minority, replace=True,n_samples=len(df_majority),random_state=52)
```

```python
df_minority_updated
```

```python
df = pd.concat([df_majority, df_minority_updated])
df['Target'].value_counts()
```

### Down Sampling


```python
import numpy as np
import pandas as pd

n_sample = 10000
class_ratio = 0.9
n_class_0 = int(n_sample * class_0_ratio)
n_class_1 = int(n_sample - n_class_0)

n_class_0, n_class_1 
```

```python
# Craeting Data Frame

class_0 = pd.DataFrame({
    'Feature 1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'Feature 2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'Target': [0] * n_class_0
})

class_1 = pd.DataFrame({
    'Feature 1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'Feature 2': np.random.normal(loc=2, scale=1, size=n_class_1),
    'Target': [1] * n_class_1
})

df = pd.concat([class_0,class_1]).reset_index(drop=True)
df
```

```python
df['Target'].value_counts()
```

```python
minority = df[df['Target'] == 1]
majority = df[df['Target'] == 0]
```

```python
from sklearn.utils import resample
df_majority_updated = resample(majority, replace=False, n_samples=len(minority), random_state=40)
```

```python
df = pd.concat([df_majority_updated, minority])
df['Target'].value_counts()
```
