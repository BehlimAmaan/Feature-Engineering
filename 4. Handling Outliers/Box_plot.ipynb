```python
import seaborn as sns
import numpy as np
```

#### 5 Number Summary


```python
lst_marks = [30, 22, 60, 55, 68, -22, -150,45, 32, 78,-30, 82, 87, 180, 95, 93, 41, 150]
min,Q1,Median,Q3,max=np.quantile(lst_marks,[0,0.25,0.50,0.75,1.0])

print(f"Minimum: {min}\nQuartile 1: {Q1}\nMedian: {Median}\nQuartile 3: {Q3}\nMaxinum: {max}")

IQR = Q3-Q1
print(f"IQR is : {IQR}")

Lower_fence = Q1-1.5*(IQR)
Higher_fence = Q3+1.5*(IQR)

print(f"Lower Fence: {Lower_fence}\nHigher Fence: {Higher_fence}")
    
```
