```python
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = sns.load_dataset('tips')
df
```

```python
df.head(10)
```

```python
encoder = OneHotEncoder()

emcide=encoder.fit_transform(df[['sex','smoker','day','time']]).toarray()
```

```python
df = pd.DataFrame(emcide,columns=encoder.get_feature_names_out())
df
```
