```python
import seaborn as sns
import numpy as np 
import pandas as np
from sklearn.preprocessing import OneHotEncoder
```

```python
df = sns.load_dataset('titanic')
df.head(10) 
```

```python
encoder = OneHotEncoder(categories=[['Third','Second','First']])
encoded = encoder.fit_transform(df[['class']])
```

```python
import pandas as pd
df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
df
```
